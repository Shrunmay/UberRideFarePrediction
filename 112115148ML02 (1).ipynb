```python
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import geopy
```

```python
def reverse_geocode(latitude, longitude):
    # Initialize the geocoder
    geolocator = Nominatim(user_agent="reverse_geocode")

    # Combine latitude and longitude into a single string
    location = f"{latitude}, {longitude}"

    try:
        # Perform reverse geocoding
        location_info = geolocator.reverse(location)
        return location_info.address
    except GeocoderTimedOut:
        # Handle timeout errors if necessary
        return None
```

```python
df = df[(df.pickup_latitude<90) & (df.dropoff_latitude<90) &
        (df.pickup_latitude>-90) & (df.dropoff_latitude>-90) &
        (df.pickup_longitude<180) & (df.dropoff_longitude<180) &
        (df.pickup_longitude>-180) & (df.dropoff_longitude>-180)]

df.head()
```

```python
from geopy.distance import geodesic
```

```python
def calculate_distance(lat1, lon1, lat2, lon2):
    # Create tuples for the two sets of coordinates
    loc1 = (lat1, lon1)
    loc2 = (lat2, lon2)

    # Calculate the distance using geodesic from the geopy library
    distance = geodesic(loc1, loc2).kilometers

    return distance
```

```python
df['distance_km'] = df.apply(lambda row: calculate_distance(row['pickup_latitude'], row['pickup_longitude'],
                                                           row['dropoff_latitude'], row['dropoff_longitude']), axis=1)
```

# Detecting and Removing Outliers Using Z-Score

```python
from scipy import stats
import numpy as np
z_scores = np.abs(stats.zscore(df))
threshold = 3
outlier_indices = np.where(z_scores > threshold)

df = df[(z_scores < threshold).all(axis=1)]

df.head()
```

# Applying Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score


X = df.drop('fare_amount', axis=1)
y = df['fare_amount']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```
